Xóa đi 2 cột không cần thiết: index_, discount_index
<br> Để thuận tiện hơn trong việc đánh xử lí và đánh giá:
<br>Đổi tên các cột: '**tittle**' thành '**name**', '**Lượt bán**' thành '**quantity_sale**', '**ram**' thành '**ram (GB)**', '**rom**' thành '**rom (GB)**', '**price**' thành '**price (đ)**', '**original price**' thành '**original price (đ)**'


```python
df1.drop(columns=['index_', 'discount_index'], inplace=True)
df1.rename(columns={'title': 'name', 'Lượt bán': 'quantity_sale', 'ram': 'ram (GB)', 'rom': 'rom (GB)', 'price': 'price (đ)', 'original price': 'original price (đ)'}, inplace=True)
df1
```

Xử lý cột 'ram (GB)' ban đầu từ kiểu object có dạng 'RAM xx GB' sang kiểu int có dạng 'xx' (đơn vị GB)
<br>Sử dụng hàm .str.extract() và Regex (\d+) để lấy các số liên tiếp trong chuỗi

```python
df1['ram (GB)']=df1['ram (GB)'].str.extract('(\d+)', expand=False).astype(int)
```

Xử lý cột 'rom (GB)' ban đầu có kiểu object có dạng "SSD xxx GB" hoặc "SSD x TB" tách thành 2 cột mới là:
- Cột 'rom (GB)' có dạng "xxx" là kiểu số: biểu thị độ lớn của rom đơn vị GB (nếu trường hợp đơn vị ban đầu là TB thì sẽ chuyển về GB)
- Cột 'rom_type': để xác định loại ổ cứng. Sử dụng hàm .str.extract() và Regex (\D+) để lấy chuỗi chữ cái liên tiếp

```python
# Tách loại ROM (SSD, HDD, eMMC)
df1["rom_type"] = df1["rom (GB)"].str.extract(r"(\D+)")

# Tách dung lượng ROM (GB)
df1["rom (GB)"] = df1["rom (GB)"].replace({"1 TB": "1024 GB", "2 TB": "2048 GB"}, regex=True)
df1["rom (GB)"] = df1["rom (GB)"].str.extract(r"(\d+)")
df1["rom (GB)"] = df1["rom (GB)"].astype(int)
```

Đưa cột 'rom_type' vừa được tạo về gần với cột 'rom (GB)' để dễ quan sát

```python
# Lấy danh sách tất cả các cột
columns = list(df1.columns)

# Chuyển "rom_type" ngay sau "rom (GB)"
columns.insert(columns.index("rom (GB)") + 1, columns.pop(columns.index("rom_type")))

# Sắp xếp lại DataFrame theo thứ tự mới
df1 = df1[columns]
```

Xử lý cột 'rate' là điểm đánh giá trung bình của sản phẩm.
Thay thế các gái trị "No rate" bằng giá trị "0". Sau đó chuyển dang kiểu dữ liệu float

```python
df1['rate']=df1['rate'].replace({'No rate': '0'}, regex=True)
df1['rate']=df1['rate'].astype(float)
```

Xử lý cột 'quantity_sale' số lượt bán của sản phẩm ban đầu có kiểu objact dạng "Đã bán xk" hoặc "Đã bán x" hoặc "No quantity sale" đối với những sản phẩm chưa có lượt bán. Đưa về kiểu số

```python
import re

def clean_quantity(q):
    if "No quantity sale" in q:  # Trường hợp chưa có lượt bán
        return 0
    match = re.search(r"([\d,.]+)k?", q)  # Tìm số có thể chứa dấu '.' hoặc ',' có thể có ký tự "k"
    if match:
        num = match.group(1).replace(",", ".")  # Đổi dấu phẩy thành dấu chấm cho số thập phân
        return float(num) * 1000 if "k" in q else float(num)  # Chuyển "k" thành 1000
    return 0  # Nếu không có số, trả về 0

# Áp dụng hàm cho cột quantity_sale
df1["quantity_sale"] = df1["quantity_sale"].apply(clean_quantity)
```

Xử lý cột "price" và "original price".
Xóa đi kí tự 'đ' và đưa về dạng số

```python
import re

def clean_price(p):
    if "No price" in p or "No discount info" in p:  # Trường hợp không có giá
        return 0
    p = re.sub(r"[^\d]", "", p)  # Xóa mọi ký tự không phải số
    return int(p)  # Chuyển thành số nguyên

df1["price (đ)"] = df1["price (đ)"].apply(clean_price)

df1['original price (đ)']=df1['original price (đ)'].apply(clean_price)
```

Xử lý cột "discount percent". Bỏ đi các kí tự khác chỉ giữ lại số và đưa về kiểu số

```python
df1['discount percent']=df1['discount percent'].replace({'No discount percent': '0'}, regex=True)
df1['discount percent']=df1['discount percent'].str.extract(r"(\d+)")
df1['discount percent']=df1['discount percent'].astype(int)
```
