Xử lý cột 'size' chuyển sang kiểu số
<br>Vì có một số điện thoại gập hiện này có 2 màn hình nên cột size sẽ có vài dòng có dạng như "Chính 7.82" & Phụ 6.31", trường hợp này sẽ lấy số liệu của màn hình chính
<br>Nếu có giá trị không chuyển được sẽ là NaN

```python
import re

def extract_first_number(text):
    if pd.isna(text):
        return None
    match = re.search(r'(\d+(\.\d+)?)', text)  # Tìm nhóm số (có thể có phần thập phân)
    if match:
        return match.group(1)  # Lấy chuỗi số đầu tiên tìm thấy
    return None

# Áp dụng hàm cho cột 'Size'
df1['size'] = df1['size'].apply(extract_first_number)

df1['size'] = pd.to_numeric(df1['size'], errors='coerce')
```

Xử lý cột 'rate' là điểm đánh giá trung bình của sản phẩm.
Thay thế các gái trị "No rate" bằng giá trị "0". Sau đó chuyển dang kiểu dữ liệu float

```python
df1['rate']=df1['rate'].replace({'No rate': '0'}, regex=True)
df1['rate']=df1['rate'].astype(float)
```

Xử lý cột 'quantity_sale' số lượt bán của sản phẩm ban đầu có kiểu objact dạng "Đã bán xk" hoặc "Đã bán x" hoặc "No quantity sale" đối với những sản phẩm chưa có lượt bán. Đưa về kiểu số

```python
import re

def clean_quantity(q):
    if "No quantity sale" in q:  # Trường hợp chưa có lượt bán
        return 0
    match = re.search(r"([\d,.]+)k?", q)  # Tìm số có thể chứa dấu '.' hoặc ',' có thể có ký tự "k"
    if match:
        num = match.group(1).replace(",", ".")  # Đổi dấu phẩy thành dấu chấm cho số thập phân
        return float(num) * 1000 if "k" in q else float(num)  # Chuyển "k" thành 1000
    return 0  # Nếu không có số, trả về 0

# Áp dụng hàm cho cột quantity_sale
df1["quantity_sale"] = df1["quantity_sale"].apply(clean_quantity)
```

Trong quá trình crawl data thì đã có sự cố, đối với 2 chiếc điện thoại cục gạch thì giá trị size lại được chuyển qua cho resolution trong khi Size lại mang giá trị là NaN.
<br>Tiến hành đổi lại cho đúng

```python
# Bước 1: Tạo một mask (điều kiện lọc) cho các dòng size NaN nhưng resolution lại có dạng kích thước
mask = df1['size'].isna() & df1['resolution'].str.contains(r'^\d+(\.\d+)?', na=False)

# Bước 2: Chuyển giá trị từ resolution sang size
df1.loc[mask, 'size'] = df1.loc[mask, 'resolution']

# Bước 3: Đặt lại resolution về NaN (hoặc chuỗi trống) cho những dòng này
df1.loc[mask, 'resolution'] = "No resolution"  # hoặc ""

# Kiểm tra lại kết quả
print(df1[mask][['size', 'resolution']])
```

```python
df1.loc[df1['original price (đ)'].isna(),'original price (đ)']="No price"
df1.loc[df1['discount percent'].isna(),'discount percent']="No discount info"
```

```python
import re

def clean_price(p):
    if "No price" in p or "No discount info" in p:  # Trường hợp không có giá
        return 0
    p = re.sub(r"[^\d]", "", p)  # Xóa mọi ký tự không phải số
    return int(p)  # Chuyển thành số nguyên

df1["price (đ)"] = df1["price (đ)"].apply(clean_price)

df1['original price (đ)']=df1['original price (đ)'].apply(clean_price)
```

Xử lý cột "discount percent". Bỏ đi các kí tự khác chỉ giữ lại số và đưa về kiểu số

```python
df1['discount percent']=df1['discount percent'].replace({'No discount info': '0'}, regex=True)
df1['discount percent']=df1['discount percent'].str.extract(r"(\d+)")
df1['discount percent']=df1['discount percent'].astype(int)
```
